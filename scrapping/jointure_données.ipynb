```python
cols_poi = df2.columns
cols_poi
```

```python
cols_poi = cols_poi.drop(['CodeZone', 'Préfecture', 'Arrondissement','lat','lon', 'geocode_quality', 'nb_education_500m', 'nb_banque_500m', 'nb_sante_500m','nb_commerce_500m','nb_transport_500m','nb_environnement_500m','nb_environnement_1000m','nb_education_1000m', 'nb_sante_1000m', 'nb_commerce_1000m', 'nb_banque_1000m', 'nb_transport_1000m', 'nb_environnement_1000m', 'nb_nuisance_1000m', 'score_amenites_500m',  'score_amenites_1km', 'nb_GAB_1km','score_nuisance_500m', 'score_accessibilite',  'centralite_routiere',])
```
