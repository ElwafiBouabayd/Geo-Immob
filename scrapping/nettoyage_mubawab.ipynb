# Nettoyage et traitement Data  Mubawab 

```python
import pandas as pd
df2 = pd.read_csv('data_mubawab.csv')
df2.head()
```

### traitement des valeurs manquantes de localisation (quartier)

```python
# copier les lignes dont la localisation est manquante
df_nan = df2[df2["location"].isna()]
df_nan.shape
```

```python
#  les lignes non NaN
df_n_nan = df2.dropna(subset=['location'])
df_n_nan.shape
```

```python
df_final = pd.concat([df_nan, df_n_nan], ignore_index=True)
```

```python
df_final = df_final[(df_final['location'] != 'Casablanca') & (df_final['location'].notna())]
```

### nettoyage et normalisation de Localisation (quartier)

```python
l=df_final['location'].value_counts()
```

```python
l=l.to_dict()
```

```python
# Liste complète des quartiers de Casablanca
import unicodedata

quartiers_casablanca = [
    'Aïn Chock', 'Ain Diab', 'Aïn Sbaa', 'Aïn Sebaâ', 'Aïn Sbaa Hay Mohammadi', 'Hay Mohammadi', 'Anfa', 'Beauséjour',
    'Belvédère', 'Ben Msik', 'Bir Anzarane', 'Bourgogne', 'Californie', 'Carrières Centrales',
    'Casablanca Medina', 'Casablanca Marina', 'Centre Ville', 'Derb Sultan', 'Derb Ghalef', 'Diab Extension', 'Douar Chaki', 'Douar Lkhalouf', 'El Fida',
    'El Hank', 'El Omaria', 'El Qods', 'Al Qods', 'Essouissi', 'Extension Est',
    'Finance City', 'Fernandez', 'Gauthier', 'Glaoui', 'Gzenaya', 'Hay Chaabi',
    'Hay Chabab', 'Hay Chrifa', 'Hay El Qods', 'Hay El Wahda', 'Hay Essaada', 'Hay Hassani',
    'Hay Jawhara', 'Hay Khayma', 'Hay Noor', 'Hay Rachid', 'Hay Salama',
    'Hay Salam', 'Hay Sawaslia', 'Hay Tar', 'Hay Warda', 'Hayat',
    'Kora Chabeb', 'La Fontaine', 'La Gironde', 'Lakhyar', 'Lalla Yacout', 'Lalla Meryem',
    'Larbaa', 'Laymoune', 'Les Hôpitaux', 'Les princesses', 'Lisanga', 'Lombarde', 'Maâmoura', 'Maarif',
    'Maarouf', 'Mahajja', 'Manara', 'Mansourah', 'Maquis', 'Alsace Lorraine',
    'Marche Verte', 'Marshala', 'Mechouar', 'Medina', 'Mers Sultan',
    'Merzak', 'Messira', 'Meunier', 'Michaux Bellaire', 'Mohammedia',
    'Moqattam', 'Msallah', 'Mzalmia', "N'Gor", 'Nakhil', 'Nassim', 'Nassim 1', 'Nassim 2',
    'New Medina', 'Niamey', 'Notre-Dame', 'Oasis', 'Oulfa', 'Outarza',
    'Pagoud', 'Palmier', 'Palmiers', 'Pasteur', 'Peyronnet', 'Pommier',
    'Pond', 'Plateau', 'Quartier Bachkou', 'Rachidi', 'Regia', 'Ain Mouloud',
    'Riad', 'Riviera', 'Roches Noires', 'Rouviere', 'Routes',
    'Rousseau', 'Roux', 'Rue Paixc', 'Rue Patrice Lumumba', 'Rue Prince',
    'Ryad Abdeljalil', 'Saada', 'Salmia', 'Salonge', 'Salti',
    'San Isidro', 'Saniat', 'Saouira', 'Sarcelles', 'Sardine',
    'Sargento', 'Satuca', 'Sauveur', 'Scala', 'Sebou',
    'Settat', 'Sidi Belyout', 'Sidi Ghanem', 'Sidi Maârouf', 'Sidi Moumen',
    'Sidi Othmane', 'Sidi Semchaoui', 'Sidi Youssef Ben Ali', 'Signor', 'Solferino',
    'Sonia', 'Souissi', 'Source', 'Sportive', 'Stadium',
    'Val Fleury', 'Dar Bouazza', 'Medyouna', 'Zenata',
]

def normalize_accents(text):
    """Supprime les accents pour la comparaison"""
    return ''.join(
        c for c in unicodedata.normalize('NFD', text)
        if unicodedata.category(c) != 'Mn'
    ).lower()

quartiers_normalized_list = [(q, normalize_accents(q)) for q in quartiers_casablanca]

def find_quartier(location_text):
    """Cherche un nom de quartier dans le texte de location"""
    if pd.isna(location_text):
        return None
    
    text = str(location_text).strip()
    text_lower = normalize_accents(text)
    
    best_match = None
    best_length = 0
    
    for quartier_original, quartier_norm in quartiers_normalized_list:
        if quartier_norm in text_lower and len(quartier_norm) > best_length:
            best_match = quartier_original
            best_length = len(quartier_norm)
    
    return best_match



# Nettoyer le dictionnaire l en normalisant les clés (locations)
l_cleaned = {}

for location, count in l.items():
    # Appliquer la fonction de normalisation
    quartier_norm = find_quartier(location)
    
    if quartier_norm:  # Si on a pu identifier le quartier
        if quartier_norm in l_cleaned:
            l_cleaned[quartier_norm] += count
        else:
            l_cleaned[quartier_norm] = count

# Trier par count décroissant
l_cleaned = dict(sorted(l_cleaned.items(), key=lambda x: x[1], reverse=True))

print("📊 DICTIONNAIRE NETTOYÉ - LOCATIONS PAR QUARTIER:\n")
print(f"{'Quartier':30} | {'Nombre'}")
print("=" * 45)
for quartier, count in l_cleaned.items():
    print(f"{quartier:30} | {count:5d}")

print(f"\n✅ Total quartiers: {len(l_cleaned)}")
print(f"✅ Total locations: {sum(l_cleaned.values())}")
```

```python
# nv df: df_final_clean en appliquant la normalisation à df_final

# Ajouter la colonne location_normalized à df_final
df_final['location_normalized'] = df_final['location'].apply(find_quartier)

# Filtrer pour garder seulement les locations reconnues
df_final_clean = df_final[df_final['location_normalized'].notna()].copy()

print(f"📊 STATISTIQUES DU NETTOYAGE:")
print(f"{'='*50}")
print(f"df_final (avant)              : {len(df_final):5d} lignes")
print(f"df_final_clean (après)        : {len(df_final_clean):5d} lignes")
print(f"Observations reconnues        : {(len(df_final_clean)/len(df_final)*100):5.1f}%")
print(f"Observations non reconnues    : {(len(df_final)-len(df_final_clean)):5d} lignes")
print(f"\n✅ df_final_clean est prêt !")
```

### Extraction du type de bien immobilier

```python
# Examiner quelques exemples de raw_text pour voir les patterns
print("🔍 EXEMPLES DE RAW_TEXT:\n")
for i, raw_text in enumerate(df_final_clean['raw_text'].head(5)):
    print(f"{i+1}. {raw_text[:150]}...")
    print()
```

```python
# Fonction pour extraire le type de bien immobilier
def extract_property_type(raw_text):
    """
    Extrait le type de bien immobilier depuis le raw_text
    Cherche les patterns: Appartement, Villa, Studio, Riad, Duplex, etc.
    """
    if pd.isna(raw_text):
        return None
    
    text = str(raw_text).lower()
    
    # Dictionnaire de patterns avec synonymes
    type_patterns = {
        'Appartement': ['appartement', 'appart', 'app '],
        'Villa': ['villa'],
        'Studio': ['studio'],
        'Riad': ['riad'],
        'Duplex': ['duplex'],
        'Penthouse': ['penthouse'],
        'Maison': ['maison'],
        'Chalet': ['chalet'],
        'Townhouse': ['townhouse', 'town house'],
        'Loft': ['loft'],
        'Mezzanine': ['mezzanine'],
        'Bungalow': ['bungalow'],
        'Résidence': ['résidence', 'residence'],
    }
    
    # Chercher les types dans le texte (priorité au premier trouvé)
    for property_type, patterns in type_patterns.items():
        for pattern in patterns:
            if pattern in text:
                return property_type
    
    return None

# Appliquer la fonction
df_final_clean['property_type'] = df_final_clean['raw_text'].apply(extract_property_type)

# Afficher les résultats
print("📊 TYPES DE BIEN IMMOBILIER DÉTECTÉS:\n")
property_counts = df_final_clean['property_type'].value_counts()
print(f"{'Type':20} | {'Nombre':6} | {'Pourcentage':10}")
print("=" * 40)
for ptype, count in property_counts.items():
    pct = (count / len(df_final_clean)) * 100
    print(f"{str(ptype):20} | {count:6d} | {pct:6.1f}%")

# Afficher les observations non reconnues
non_recognized = df_final_clean[df_final_clean['property_type'].isna()]
print(f"\n⚠️  Non reconnus: {len(non_recognized)} ({(len(non_recognized)/len(df_final_clean)*100):.1f}%)")
```

```python
# Fonction améliorée pour extraire le type de bien immobilier
def extract_property_type_v2(raw_text):
    """
    Extraire le type de bien immobilier avec meilleure couverture
    """
    if pd.isna(raw_text):
        return None
    
    text = str(raw_text).lower()
    
    # Dictionnaire hiérarchisé par priorité (du plus spécifique au plus général)
    type_patterns = {
        'Penthouse': ['penthouse'],
        'Townhouse': ['townhouse', 'town house'],
        'Duplex': ['duplex'],
        'Riad': ['riad'],
        'Chalet': ['chalet'],
        'Mezzanine': ['mezzanine'],
        'Loft': ['loft'],
        'Studio': ['studio'],
        'Villa': ['villa'],
        'Maison': ['maison'],
        'Appartement': ['appartement', 'appart ', 'app '],
        'Immeuble': ['immeuble'],
        'Bureau': ['bureau', 'bureaux', 'plateau'],
        'Local commercial': ['local commercial', 'locaux commercial'],
        'Résidence': ['résidence', 'residence'],
        'Bungalow': ['bungalow'],
    }
    
    # Chercher les types (première correspondance = priorité)
    for property_type, patterns in type_patterns.items():
        for pattern in patterns:
            if pattern in text:
                return property_type
    
    return None
```
